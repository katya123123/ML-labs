## Обработка пропусков

```python
df['Mjob'][350:] = None
df['absences'][:150] = None
df
```

```python
df_cleaned = df.copy()
df_cleaned = df.dropna()
df_cleaned
```

Среднее значение показателя пропусков довольно значительно изменилось по сравнению с изначальными значениями, поскольку мы "потеряли" часть значений. Изменилось, хоть и не так значительно, среднеквадратичное отклонение, оно стало больше, что означает, что значения стали более "разбросаны" в своём интервале.

```python
df_filled_by_mean = df.copy()
df_filled_by_mean['absences'] = df['absences'].fillna(df['absences'].mean())
df_filled_by_mean
```

```python
df_filled_by_mean.describe()
```

Заполнение пропуска средним значением также привел к значительному изменению среднего, однако в данном случае очень сильно уменьшилось среднеквадратичное отклонение. Так произошло по причине того, что огромная часть значений теперь являются средним значением, соответствие и разброс значений стал куда меньше относительно среднего. При этом стоит отметить, что значительно измениля 2 и 3 квартили. На мой взгляд, такие данные уже не подходят для анализа.

```python
df_filled_by_median = df.copy()
df_filled_by_median['absences'] = df['absences'].fillna(df['absences'].median())
df_filled_by_median
```

```python
df_filled_by_median.describe()
```

При заполнении пропуском медианой мы видим, что изменение среднего незначительно и при этом среднеквадратичное отклонение значительно уменьшилось. Данные при таком способе заполнения пропуска притерпели минимальные изменения.

```python
print(df['absences'].mean(), df['absences'].median())
```

Заполнение значений необходимо, чтобы не терять строки с нулл-значениями, что происходит при обычном удалении нулл-значений. Самыми простыми способами заполнения значений являются заполнение на основе среднего и медианы.

В данном случае, медиана значительно отличается от среднего, за счет чего оба способа дают различный рельтат. Первый способ скорее подошел бы, если бы эта колонка не влияла на результат итоговых вычислений и анализа данных, а второй способ подойдет, если значение колонки влияет на конечный результат и нам важно сохранить корректность данных, но при этом заполнить пропуски.

```python
df_filled_by_moda  = df.copy()
df_filled_by_moda['Mjob'] = df['Mjob'].fillna(df['Mjob'].mode()[0])
df_filled_by_moda
```

Так как пропуски встречаются в том числе и в колонке Mjob, которая является категориальной, её пустые значения необходимо заполнить на основе моды. Значения заполнились значением other, поскольку это самое часто встречающееся слово в данной колонке.


```python
df_interpolated = df.copy()
for i in range(len(df_interpolated['absences'])):
  if df_interpolated['absences'][i] == None:
    continue
  else:
    df_interpolated['absences'][i] = df_interpolated['absences'][i].astype('int')
df_interpolated['absences'] = df_interpolated['absences'].interpolate()
df_interpolated
```

```python
df.describe()
```

Данный метод работет крайне криво за счет того, что у нас много подряд пустых значений, а они не разбросаны как единичные случаи по колонке. Таким образом, данный способ подходит только в том случае, когда нет множества подряд пропущенных значений.
