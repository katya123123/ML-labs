```python
#Звездочка
import numpy as np
import csv
```

```python
with open('student-mat.csv') as f:
  arr = list(csv.reader(f, delimiter=','))

arr = np.array(arr)
arr
```

```python
category = arr[:,8]
cat_clr = np.delete(category, 0, 0)
```

```python
num = arr[:,2]
num_clr = np.delete(num, 0, 0).astype(int)
```

```python
at_home = np.array([])
service = np.array([])
other = np.array([])
teacher = np.array([])
health = np.array([])
for i in range(len(cat_clr)):
  if cat_clr[i] == 'at_home':
    at_home = np.insert(at_home, 0, num_clr[i])
  elif cat_clr[i] == 'services':
    service = np.insert(service, 0, num_clr[i])
  elif cat_clr[i] == 'other':
    other = np.insert(other, 0, num_clr[i])
  elif  cat_clr[i]== 'teacher':
    teacher = np.insert(teacher, 0, num_clr[i])
  elif cat_clr[i] == 'health':
    health = np.insert(health, 0, num[i])

print('at_home mean: ', np.mean(at_home.astype(int)), '\n',
      'service mean: ', np.mean(service.astype(int)), '\n',
      'other: ', np.mean(other.astype(int)), '\n',
      'teacher mean: ', np.mean(teacher.astype(int)), '\n',
      'health: ', np.mean(health.astype(int)), '\n')

print('at_home median: ', np.median(at_home.astype(int)), '\n',
      'service median: ', np.median(service.astype(int)), '\n',
      'other:  median: ', np.median(other.astype(int)), '\n',
      'teacher median: ', np.median(teacher.astype(int)), '\n',
      'health  median: ', np.median(health.astype(int)), '\n')

```
